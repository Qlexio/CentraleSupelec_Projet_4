À première vue, je vois qu'il est possible de réduire les individus d'une commande en fonction du produit mais aussi en fonction de la catégorie et pondérer l'article sur la commande. Cela devrait me permettre de limiter les doublons de même que la pondération peut permettre d'estimer l'impact des produits sur les avis.

Le but est de trouver une formule prenant en compte le prix des produits et leur quantité.

Hypothèse de formule pour la pondération:
* utilisation de la fonction cumulative de la distribution normale $f(PX)$ sur les prix uniques $PX = px_1, px_2, \dots, px_n$
  * $f(PX) \in R^+$ avec:
    * $f(PX) = Norm(\nu, \frac{\nu}{3})$ sur $(0, \nu)$
    * $f(PX) = Norm(\nu, \sigma)$ sur $(\nu, +\infty)$
* calcul de la proportion de chaque produit par rapport au nombre total de produits de la commande => $PC = pc_1,pc_2, \dots,pc_n$
* Le poids de chaque produit est calculé par:
    $W = \displaystyle \frac{(f(px_i) + 0.5) * (pc_i + 0.5)}{\sum_{i=1}^{n} (f(px_i) + 0.5) * (pc_i + 0.5)}$ 

```python
import pprint

def calc(prix):
    nu = np.mean(prix)
    sigma = np.std(prix, ddof=0)
    
    px_unique = sorted(set(prix))
    p_norm = {}
    if len(px_unique) > 1:
        for n in px_unique:
            if n < nu:
                p_norm[n] = np.round(norm.cdf(n, loc= nu, scale= nu/3), 5) + 0.5
            else:
                p_norm[n] = np.round(norm.cdf(n, loc= nu, scale= sigma), 5) + 0.5
    else:
        p_norm[px_unique[0]] = 1

    nb_prix_diff = {i: prix.count(i) for i in prix}
    
    ### H5 (+0.5, +0.5)
    perc = {}
    for k, v in nb_prix_diff.items():
        perc[k] = np.round(v / len(prix), 5) + 0.5

    tmp_calc = {}
    for px in px_unique:
        tmp_calc[px] = np.round(p_norm[px] * perc[px], 5)

    somme = np.round(sum(list(tmp_calc.values())), 5)
    resultH5 = {}
    for px in px_unique:
        resultH5[px] = {}
        resultH5[px]["nb article"] = nb_prix_diff[px]
        resultH5[px]["poids"] = np.round(tmp_calc[px] / somme, 5)
        resultH5[px]["poids/article"] = np.round(resultH5[px]["poids"] / nb_prix_diff[px], 5)

    print("H5 (+0.5, +0.5)")
    pprint.pprint(resultH5)
```
