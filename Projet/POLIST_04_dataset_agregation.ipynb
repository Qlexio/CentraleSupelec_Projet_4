```python
from scipy.stats import norm

def weight(data, sort_col, prod_col, price_col):
    data_values = data[sort_col].value_counts().reset_index()

    results = []
    
    for i in range(len(data_values)):
        tmp_datas = data.loc[data[sort_col] == data_values.iloc[i, 0], :]

        prices = tmp_datas[price_col].tolist()
        nu = np.mean(prices)
        sigma = np.std(prices, ddof=0)
    
        px_unique = tmp_datas.groupby([sort_col, prod_col])[[sort_col, prod_col, price_col]].mean().round(2).reset_index().to_numpy().tolist()
        px_count = tmp_datas.groupby(prod_col)[[price_col]].count().reset_index().to_numpy().tolist()

        p_norm = {}
        if sigma != 0:
            for n in px_unique:
                if n[2] < nu:
                    p_norm[n[1]] = np.round(norm.cdf(n[2], loc= nu, scale= nu/3), 5) + 0.5
                else:
                    p_norm[n[1]] = np.round(norm.cdf(n[2], loc= nu, scale= sigma), 5) + 0.5
        else:
            for m in range(len(px_unique)):
                p_norm[px_unique[m][1]] = 1


        perc = {}
        for c in px_count:
            perc[c[0]] = np.round(c[1] / len(prices), 5) + 0.5

        tmp_calc = {}
        for px in px_unique:
            tmp_calc[px[1]] = np.round(p_norm[px[1]] * perc[px[1]], 5)

        somme = np.round(sum(list(tmp_calc.values())), 5)

        for px in px_unique:
            results.append([px[0], px[1], np.round(tmp_calc[px[1]] / somme, 5)])

    return results
```

### 2. Autres features

```python
cols = ["price", "freight_value", "product_description_lenght", "product_photos_qty", "weight", "nb_product"]
cols_excl = ["weight", "product_description_lenght", "product_photos_qty"]
type_val = ["min", "mean", "max", "sum"]
```

```python
def new_feat(data, cols, cols_excl, sort_col, type_val):
    # extraire sort_col uniques ("order_id")
    sort_values = data[sort_col].value_counts().reset_index()

    # créer colonnes associées à sort_col + chaque cols
    final_datas = {}
    final_datas[sort_col] = []
    for col in cols:
        for type in type_val:
            if col not in cols_excl or type != "sum":
                final_datas[col + "_" + type] = []

    # boucler sur chaque résultat
    for i in range(len(sort_values)):
        # récupérer les individus de la commande par sort_col
        val = sort_values.iloc[i, 0]
        tmp_datas = data.loc[data[sort_col] == val, cols]
        final_datas[sort_col].append(val)

        # boucler sur cols et ajout de la valeur dans la colonne associée
        for col in cols:
            # min
            final_datas[col + "_min"].append(tmp_datas[col].min())
            # max
            final_datas[col + "_max"].append(tmp_datas[col].max())
            # somme sauf si cols == cols_excl
            if col not in cols_excl:
                if col in cols[-1:]:
                    final_datas[col + "_sum"].append(np.round(tmp_datas[col].sum(), 2))
                else:
                    final_datas[col + "_sum"].append(np.sum(tmp_datas[col] * tmp_datas[cols[-1]]))
            # moyenne
            if col in cols[:2]:
                final_datas[col + "_mean"].append(np.round(np.sum(tmp_datas[col] * tmp_datas[cols[-1]]) / tmp_datas[cols[-1]].sum(), 2))
            else:
                final_datas[col + "_mean"].append(np.round(tmp_datas[col].mean(), 2))

    return pd.DataFrame(final_datas)
    
```
