```python
import POLIST_00_graph_analysis as ga
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

datas = pd.read_csv("POLIST_13_kmean_datas.csv")
```

```python
from sklearn.preprocessing import OneHotEncoder

X = datas.customer_state.to_numpy().reshape(-1, 1)
ohe = OneHotEncoder().fit(X)

X_ohe = pd.DataFrame(ohe.transform(X).toarray(), columns= ohe.get_feature_names_out(["customer_state__"]), index= datas.index)
```

```python
datas.drop("customer_state", axis= 1, inplace= True)

datas = datas.merge(X_ohe, left_index= True, right_index= True)
datas
```

```python
from POLIST_00_graph_analysis import segmentation

segment_names = ["cluster " + str(x) for x in np.sort(datas["labels"].unique())]
conditions = [datas['labels'] == x for x in np.sort(datas["labels"].unique())]

segments = segmentation(datas, conditions, segment_names)
segments
```
